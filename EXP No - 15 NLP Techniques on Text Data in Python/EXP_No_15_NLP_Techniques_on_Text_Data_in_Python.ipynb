```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

text = "Natural language processing is interesting"

tokens = word_tokenize(text)

print(tokens)
```

```python
# Sentence Tokenization
from nltk.tokenize import sent_tokenize

text = "Python is easy. It is widely used in data science"
sentence = sent_tokenize(text)

print(sentence)
```

```python
# Stop word removal
# Stop words such as is , the,and,in are removed

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))

text = "Python is easy. It is widely used in data science"
words = word_tokenize(text)


filtered_words = [word for word in words if word.lower() not in stop_words]

print(filtered_words)
```

```python
# Stemming

# Stemming reduces words to their root form

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "runs", "ran","playing","watched","studies","coding"]

for w in words:
  stemmed_word = stemmer.stem(w)
  print(stemmed_word)

```

```python
# Lemmatization
# Lemmatization converts words to their dictionary form.

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running"))
print(lemmatizer.lemmatize("runs"))
print(lemmatizer.lemmatize("ran"))
print(lemmatizer.lemmatize("playing"))
print(lemmatizer.lemmatize("watched"))
print(lemmatizer.lemmatize("studies"))
print(lemmatizer.lemmatize("coding"))
```

**POS Tag Definitions**

NNP : Proper noun, singular (e.g., "London", "Microsoft")

NN : Noun, singular or mass (e.g., "bicycle", "air")

NNS : Noun, plural (e.g., "bicycles", "mountains")

VBZ : Verb, 3rd person singular present (e.g., "writes", "takes")

VBP : Verb, non-3rd person singular present (e.g., "write", "take")

VBD : Verb, past tense (e.g., "wrote", "took")

VBG : Verb, gerund or present participle (e.g., "writing", "taking")

VBN : Verb, past participle (e.g., "written", "taken")

JJ : Adjective (e.g., "green", "large")

JJR : Adjective, comparative (e.g., "greener", "larger")

JJS : Adjective, superlative (e.g., "greenest", "largest")

RB : Adverb (e.g., "quickly", "extremely")

DT : Determiner (e.g., "the", "a", "an")

PRP : Personal pronoun (e.g., "I", "he", "it")

PRP$ : Possessive pronoun (e.g., "my", "his", "its")

IN : Preposition or subordinating conjunction (e.g., "in", "of", "during")

CC : Coordinating conjunction (e.g., "and", "but", "or")

CD : Cardinal number (e.g., "one", "2026")

MD : Modal (e.g., "can", "should", "will")
