```python
'''4.Temprature Analysis'''
temperature=[26,25,29,27,28,21,22,18,19,23]
print("The tempratures of first 5 days")
print(temperature[:5])
print("The temprature of last 5 days")
print(temperature[-5:])
print("Highest temprature")
print(max(temperature))
print("Lowest temprature")
print(min(temperature))
print("Average temprature")
print(sum(temperature)/len(temperature))
```

```python
'''5.Daily expense tracker'''
expense_tracker=[170,250,130,125,300,225,160]
print("Total expense")
print(sum(expense_tracker))
print("Average expense")
print(sum(expense_tracker)/len(expense_tracker))
print("Highest spending day")
print(max(expense_tracker))
print("Lowest spending day")
print(min(expense_tracker))
```
