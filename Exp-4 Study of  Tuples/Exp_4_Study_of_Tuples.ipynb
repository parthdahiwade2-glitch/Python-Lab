```python
# Slicing
numbers =(3,7,1,9,4)
print(numbers[1::2])
```

```python
numbers = (10,20,30,40,50)
# Slicing examples
print("Elements from 1:3")
print(numbers[1:3])
print("First three elements:")
print(numbers[:3])
print("Elements from index 2 onwards:(Reverse Indexing)")
print(numbers[-3:])
print("All elements:")
print(numbers[:])
```
