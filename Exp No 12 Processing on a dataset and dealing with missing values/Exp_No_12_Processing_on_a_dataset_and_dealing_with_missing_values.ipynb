```python
# Processing on a dataset - 1
import pandas as pd

data = {
    "Roll No": [101,102,103,104,105],
    "Name": ["John", "Emily", "Michael", "Sarah", "David"],
    "Age": [20,"-",21,20,"22"],
    "Marks":[85,90,"-",88,76],
    "Department":["CSE","IT","CSE","cse","E&TC"],
    "Admission_Date":["2022-01-15","03-06-2023","05-06-2023","07-06-2023","10/06/23"]

}

df = pd.DataFrame(data)

# Save as CSV

df.to_csv("student_data.csv", index=False)



```

```python
import pandas as pd
import numpy as np
# Step - 1 Load the dataset

df = pd.read_csv("student_data.csv")

print("Original Data:")
print(df)
```

```python
df.replace("-",np.nan,inplace=True)
df
```

```python
# Step - 2 '-' With NaN to mark missing values

df.replace("-",np.nan, inplace=True)

# Convert columns to numeric where needed
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Marks"] = pd.to_numeric(df["Marks"], errors="coerce")

''' When converting data (like strings to datatime),
if some values are invalid or cannot be converted,
insead of throwing an error, it will:
Replace those invalid values with NaT (Not a Time)
'''
df
```

```python
# Step - 3 Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Fill the missing age with mean
df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
# Fill missing marks with median

df["Marks"].fillna(df["Marks"].median(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Step 5 Fix inconsistent text formattig
df["Department"] = df["Department"].str.upper()
```

```python
# Step 6 Convert Admission_Date to standard date time format
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"], errors='coerce')
```

```python
df
```

```python
# Step 7 Show cleaned data
print("\nCleaned Data:")
print(df)
```

```python
# Step 8 Save cleaned dataset
df.to_csv("cleaned_student_data.csv",index=False)

print("\nCleaned CSV file saved as cleaned_student_data.csv")
```

```python
# Working with missing values on dataset-2
df=pd.read_csv("/Cars93 (2).csv")
df
```

```python
df.isna().sum()
```

```python
# Imputing the missing value
#1. Luggage Room
df['Luggage.room']=df['Luggage.room'].fillna(df['Luggage.room'].mean())
```

```python
# Rear.seat.room
df["Rear.seat.room"]=df["Rear.seat.room"].fillna(df["Rear.seat.room"].mean())
```

```python
# .AirBags
df['AirBags']=df['AirBags'].fillna(value=df['AirBags'].mode().values[0])
```

```python
# Save cleaned dataset
df.to_csv("cleaned_Cars93.csv",index=False)
```
