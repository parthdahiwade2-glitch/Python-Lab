```python
# Dataset

import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],

    'Price':['$800','$200','$300','$100','$250','$50'],
    'Units_Sold':[100,50,75,20,30,40],
    'Discount':[10,15,12,20,18,8]
}
df = pd.DataFrame(data)
print(df)
```

```python
# 1. Min Max Normalisation for Price
df['Price'] = df['Price'].replace({r'[$,]': ''}, regex=True).astype(float)
df['Price_MinMax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())
print("\n Min-Max Normalisation for Price")
print(df[['Product' , 'Price', 'Price_MinMax']])
```

```python
# Normalisation for for multiple columns

cols = ['Price', 'Units_Sold', 'Discount']

df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())

print("\nNormalized Dataset")
print(df_norm)
```

```python
# Z-score Normalization
df['Units_Zscore'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()

print("\nZ-score Normalization for Units_Sold")
print(df[['Product', 'Units_Sold', 'Units_Zscore']])
```

```python
# 3. Decimal Scaling
df['Price_Decimal'] = df['Price'] / 100000

print("\nDecimal Scaling for Price")
print(df[['Product', 'Price', 'Price_Decimal']])
```

## Turning catogarical variables into quantitive variables in Python

```python
import pandas as pd
import numpy as np

data = {
    'Order_ID':[101,102,103,104,105,106],
    'Customer_Gender':['Male','Female','Female','Male','Female','Male'] ,
    'Payment_Method':['Credit Card','PayPal','PayPal','Credit Card','Cash on Delivery','Credit Card'],
    'Product_Category':['Electronics','Clothing','Electronics','Clothing','Electronics','Clothing'],
    'City':['Mumbai','Delhi','Mumbai','Delhi','Mumbai','Delhi'],
    'Order_Value':[1000,500,800,1200,1500,900],
}
df = pd.DataFrame(data)
print(df)
```

```python
# Method 1: Label E ncoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Customer_Gender_Encoded'] = le.fit_transform(df['Customer_Gender'])

print(df)
```

```python
# Method 2: One hot encoding

df_encoded = pd.get_dummies(df, columns=['Payment_Method'])

print(df_encoded)
```

```python
# Method 3: Dummy Encoding(Drop First)

df_dummy = pd.get_dummies(df, columns=['Payment_Method'], drop_first=True)

print(df_dummy)
```
