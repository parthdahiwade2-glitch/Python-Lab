```python
''' Find out wether a given year is leap year or not .'''

year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
 ''' Write a python program to check if a date is valid and print the incremented date if it is valid.'''

date_str = input("Enter date in format dd/mm/yyyy: ")
try:
    dd, mm, yyyy = map(int, date_str.split('/'))
except ValueError:
    print("Invalid date format. Please use dd/mm/yyyy.")
    exit()

if not (1 <= mm <= 12):
    print("Invalid month.")
    exit()

max_days = 0

if mm in (1, 3, 5, 7, 8, 10, 12):
    max_days = 31
elif mm in (4, 6, 9, 11):
    max_days = 30
elif mm == 2:
    if (yyyy % 4 == 0 and yyyy % 100 != 0) or (yyyy % 400 == 0):
        max_days = 29
    else:
        max_days = 28

if not (1 <= dd <= max_days):
    print(f"Invalid day for month {mm} in year {yyyy}.")
    exit()


print(f"The entered date {date_str} is valid.")

dd += 1

if dd > max_days:
    dd = 1
    mm += 1
    if mm > 12:

        mm = 1
        yyyy += 1

print(f"Incremented date: {dd:02d}/{mm:02d}/{yyyy}")
```

```python
''' Write a python program to calculate Income tax based on annual income:

Up ti Rs.2,50,000: No tax
Rs.2,50,001-Rs.5,00,000 : 5%
Rs.5,00,001 - Rs.10,00,000: 20%

Above Rs.10,00,000 :30% '''

annual_income = float(input("Enter your annual income: "))
income_tax = 0

if annual_income <= 250000:
    income_tax = 0
elif annual_income <= 500000:
    income_tax = (annual_income - 250000) * 0.05
elif annual_income <= 1000000:
    income_tax = (250000 * 0.05) + (annual_income - 500000) * 0.20
else:
    income_tax = (250000 * 0.05) + (500000 * 0.20) + (annual_income - 1000000) * 0.30

print(f"Your annual income is: Rs. {annual_income:,.2f}")
print(f"Your calculated income tax is: Rs. {income_tax:,.2f}")
```

```python
''' Write a python program to calculate the gross salary of an employee based on basic salary.
If basic <= 10000: HRA = 20%, DA = 80%

If basic <= 20000: HRA = 25 %, DA = 90%

Otherwise : HRA = 30%, DA = 95%'''

basic= float(input("Enter basic salary:"))

if basic <= 10000:
    hra = 0.20 * basic
    da = 0.80 * basic
elif basic <= 20000:
    hra = 0.25 * basic
    da = 0.90 * basic
else:
    hra = 0.30 * basic
    da = 0.95 * basic

gross_salary = basic + hra + da

print(f"Basic Salary: {basic:.2f}")
print(f"HRA: {hra:.2f}")
print(f"DA: {da:.2f}")
print(f"Gross Salary: {gross_salary:.2f}")
```
