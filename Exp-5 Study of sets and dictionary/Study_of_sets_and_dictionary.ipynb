```python
''' Three students choose elective subjects .
Find the common subjects chosen by all students using set operations .
Concept used: Set Intersections (&)'''

Anna_subjects = {"Physics", "Chemistry","Enviromental studies", "Math", "Computer Science"}
Mike_subjects = {"Chemistry", "Biology", "Math", "English","Enviromental studies"}
Chris_subjects = {"Math", "Physics","Enviromental studies", "Computer Science", "History"}

common_subjects = Anna_subjects & Mike_subjects & Chris_subjects

print("Common subjects chosen by all students:", common_subjects)
```

```python
''' One set contains students in Cricket Club , another in the Fotball Club.
Find students who are:1. In both clubs 2. In only one club
Concept used: Intersections, Symmetric difference'''

cricket_club = {"Alice", "Bob", "Charlie", "David", "Eve"}
football_club = {"Bob", "David", "Frank", "Grace", "Henry"}

students_in_both_clubs = cricket_club & football_club
print("Students in both clubs:", students_in_both_clubs)

students_in_only_one_club = cricket_club ^ football_club
print("Students in only one club:", students_in_only_one_club)
```

```python
''' Absent students
Given a set of all students and a set of present students , find the absent students.
Concept used: Set Difference'''

all_students = {"Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Henry","Chris"}
present_students = {"Alice", "Bob", "Charlie", "David", "Eve"}

absent_students = all_students - present_students
print("Absent students:", absent_students)
```

```python
''' A dictionary stores user names and passwords . Write a python program to validate user login.'''

user_credentials = {
    "Alice":"alice@12345",
    "Anik":"anik@12345",
    "Krish":"krish@12345",
    "Pushpak":"pushpak@12345",
    "Bhavya":"bhavya@12345"
}

print("User Login")

username = input("Enter your username: ")

password = input("Enter your password: ")

if username in user_credentials and user_credentials[username] == password:
    print("Login successful!")
else:
    print("Invalid username or password.")
```

```python
''' A dictionary stores student names and their marks.
Write a python program to find the student who has scored the highest marks'''

student_marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "Mason": 67,
}

highest_marks_student = max(student_marks, key=student_marks.get)

print("Student with the highest marks:", highest_marks_student)

print("Marks:", student_marks[highest_marks_student])
```
